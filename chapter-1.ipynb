Use function notation to express the weight of a pig in pounds as a function of its age in days

```python
## Pigs grow 1.5kg per day
growth = 1.5
days = [50, 70, 80]

def weight_as_function_of_age(days):
    # enumerate gives us the index (starting at 0) and the value
    for index, day in enumerate(days):
        weight = day * growth
        # In Python, we use print() and f-strings start with f""
        print(f"pig number {index} has been growing for {day} days and so weighs {weight} Kilograms")

weight_as_function_of_age(days)

```

```python
growth = 1.5
days = [50, 70, 80]

def seasonal_weight_function(days):
    for index, day in enumerate(days):
        # math.sin(day) oscillates between -1 and 1.
        # This simulates the pig eating more or less depending on the 'season'
        seasonal_factor = math.sin(day) 
        weight = day * (growth + seasonal_factor)
        
        print(f"pig number {index}:")
        print(f"  - Age: {day} days")
        print(f"  - Seasonal Factor: {seasonal_factor:.2f}")
        print(f"  - Total Weight: {weight:.2f} kg\n")

seasonal_weight_function(days)
```

```python
# 1. Setup Data
# We use np.linspace to create 100 smooth points between day 0 and 100
days = np.linspace(0, 100, 500) 
growth_rate = 1.5

# 2. Define the Mathematical Functions
linear_growth = days * growth_rate
# We multiply the sine wave by 10 to make the 'wiggles' visible
seasonal_growth = days * (growth_rate + 0.2 * np.sin(days / 5))

# 3. Create the Plot
plt.figure(figsize=(10, 6))

# Plot the Linear Trend (The 'Expected' Growth)
plt.plot(days, linear_growth, label='Linear Growth (Expected)', linestyle='--', color='gray')

# Plot the Seasonal Trend (The 'Real-World' Growth)
plt.plot(days, seasonal_growth, label='Seasonal Growth (Actual)', color='blue', linewidth=2)

# 4. Add Labels (The 'Metadata')
plt.title("Pig Growth Modeling: Pre-Calculus in Action", fontsize=14)
plt.xlabel("Days Since Birth")
plt.ylabel("Weight (kg)")
plt.legend()
plt.grid(True, alpha=0.3)

# 5. Show it!
plt.show()
```

```python
# 1. Define the function (as we did before)
def h(p):
    return p**2 + 2 * p


# 2. Create an array of 'p' values (the x-axis)
# This creates 400 points between -10 and 10
p_values = np.linspace(-5, 5, 10)

# 3. Calculate the 'h(p)' values for every p (the y-axis)
h_values = h(p_values)

# 4. Create the plot
plt.figure(figsize=(8, 6))
plt.plot(p_values, h_values, label="$h(p) = p^2 + 2p$", color="blue")

plt.axhline(y=1, color="red", linestyle="--", label="y = 1")

# This creates a list of integers from -5 to 5 for the x-axis
plt.xticks(np.arange(-5, 6, 1))

# This creates a list of integers from the bottom to the top of your data
# h(-1) is -1 and h(5) is 35, so we set a reasonable range:
plt.yticks(np.arange(-2, 35, 5))

# 5. Add labels and a grid for readability
plt.title("Graph of the Function h(p)")
plt.xlabel("p (Input)")
plt.ylabel("h(p) (Output)")
plt.axhline(0, color="black", linewidth=1)  # Adds the x-axis line
plt.axvline(0, color="black", linewidth=1)  # Adds the y-axis line
plt.grid(True, linestyle="--")
plt.legend()

# 6. Show the graph
plt.show()
```

```python
def solve_for_x(y):
    """
    Solves x - 8y^3 = 0 for x.
    Rearranged: x = 8 * y^3
    """
    return 8 * (y**3)


# Example usage:
result = solve_for_x(2)
print(f"When y is 2, x is {result}")

import matplotlib.pyplot as plt
import numpy as np

# 1. Define the range for x
x = np.linspace(-20, 20, 500)

# 2. Calculate y (solving x - 8y^3 = 0 for y)
# y = (x / 8)^(1/3) = x^(1/3) / 2
# We use np.sign and np.abs to handle negative numbers properly
y = np.sign(x) * (np.abs(x) ** (1 / 3)) / 2

# 3. Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$y = \frac{\sqrt[3]{x}}{2}$", color="royalblue", linewidth=2)

# 4. Styling (making it look professional)
plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)
plt.grid(True, linestyle="--", alpha=0.6)
plt.title("Visualization of the Cubic Activation Function", fontsize=14)
plt.xlabel("Input (x)", fontsize=12)
plt.ylabel("Output (y)", fontsize=12)
plt.legend()

# 5. Display/Save
plt.savefig("cubic_root_plot.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define domains
domains = [(-0.1, 0.1), (-10, 10), (-100, 100)]

plt.figure(figsize=(12, 4))

for i, (a, b) in enumerate(domains, 1):
    x = np.linspace(a, b, 500)
    y = x**2

    plt.subplot(1, 3, i)
    plt.plot(x, y)
    plt.title(f"Domain: [{a}, {b}]")
    plt.xlabel("x")
    plt.ylabel("y = x^2")

    # Range is [0, max(x^2)]
    y_min = 0
    y_max = max(a**2, b**2)
    plt.text(
        0.5,
        0.9,
        f"Range: [{y_min}, {y_max}]",
        transform=plt.gca().transAxes,
        ha="center",
    )

plt.tight_layout()
plt.show()
```
